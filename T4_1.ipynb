```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# importando os dados utilizando o pandas
data = pd.read_csv('aug_train.csv')
```

```python
# descricao das amostragens dos dados
data.describe()
```

```python
# visualização das primeiras colunas
data.head()
```

```python
# visualização da distribuição dos valores no target
data['Response'].value_counts()
```

```python
# separando linhas com target = 0 e linhas com target = 1
classe_positiva = data[data['Response'] == 1]
classe_negativa = data[data['Response'] == 0]
```

```python
# pegando aleatoriamente exemplos do target = 0 da mesma quantidade do target =1
classe_negativa_subamostrada = classe_negativa.sample(n=len(classe_positiva), random_state=42)
```

```python
# concatenando os 2 dfs e verificando se as classes estao com quatidades iguais
data_balanceado = pd.concat([classe_positiva, classe_negativa_subamostrada])

data_balanceado['Response'].value_counts()
```

```python
# executando o one-hot encoding utilizando get_dummies do pandas
df = pd.get_dummies(data_balanceado, drop_first=True)
```

```python
# preparando para treinamento
# separando em 2 dfs, 1 com as features e outro com o target

X = df.drop(columns= 'Response')
y = df['Response']
```

```python
# dividir dados de treino e de teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Treinar o modelo piloto
modelo_piloto = LogisticRegression(random_state=42)
modelo_piloto.fit(X_train, y_train)
```

```python
y_pred = modelo_piloto.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Acurácia do modelo piloto: {accuracy}")
```
